L'utilisation de la fonction "melt" permet de convertir un DataFrame dans un format où une ou plusieurs colonnes sont des variables d'identification (id_vars), tandis que toutes les autres colonnes, considérées comme des variables mesurées (value_vars), sont "dépivotées" sur l'axe des lignes, laissant seulement deux colonnes de non-identification, "variable" et "valeur".

```python
df = table1.melt(id_vars=['maille_nom'])
```

```python
df['trimestre'] = 'T1-2020'
```

```python
df
```

```python
table2=pd.read_sql_query('SELECT * FROM T22020',engine)
```

```python
table2
```

Regroupement table 1 Q1 et table 2 Q2

```python
df2 = table2.melt(id_vars=['maille_nom'])
```

```python
df2['trimestre'] = 'T2-2020'
```

```python
df2
```

```python
pd.concat([df, df2], axis=0)
```

Même action cette fois avec la table 3 contenant les données du troisième trimestre. Cette action est a répétée avec les tables restantes jusqu'à temps de grouper l'ensemble des tables. Nous avons dû procéder de cette façon, car les données disponibles sur Pôle emploi ne nous a pas permis de récupérer les données sur la période donnée (T12020-T12021) en une seule fois. 

```python
table3=pd.read_sql_query('SELECT * FROM T32020',engine)
```

```python
df3 = table3.melt(id_vars=['maille_nom'])
```

```python
df3['trimestre'] = 'T3-2020'
```

```python
pd.concat([df, df2,df3], axis=0)
```

```python
table4=pd.read_sql_query('SELECT * FROM T42020',engine)
```

```python
df4 = table4.melt(id_vars=['maille_nom'])
```

```python
df4['trimestre'] = 'T4-2020'
```

```python
pd.concat([df, df2,df3,df4], axis=0)
```

```python
table5=pd.read_sql_query('SELECT * FROM T12021',engine)
```

```python
df5 = table5.melt(id_vars=['maille_nom'])
```

```python
df5['trimestre'] = 'T1-2021'
```

```python
pd.concat([df, df2,df3,df4,df5], axis=0)
```

Sauvegarde dans une variable des données:

```python
emploi = pd.concat([df, df2,df3,df4,df5], axis=0)
```

Une fois les données des deux tables triées, nous allons exporter les deux fichiers sous format SQL pour pouvoir faire la jointure entre les deux sources de données pour pouvoir les comparer :

```python
emploi.to_csv('Desktop/Ironhack/Projet1/offres_emploi.csv')
```
