```python
import pandas as pd
```

```python
data = pd.read_csv('../pauline/Desktop/chiffres-cles.csv', sep=";")
```

```python
data.head()
```

Les données disponibles dans le fichier sont à la fois les données concernant la France entière et les régions, c'est pourquoi il a fallu filtrer par département 

```python
df =data[data.granularite=='departement']
```

```python
df.granularite.unique()
```

```python
df["date"]=pd.to_datetime(df["date"])
```

```python
df.info()
```

```python
df.head()
```

```python
df.index=df.date
```

Ici, nous avons fait un regroupement des données par mois puis un tri alphabétique par département 

```python
df.groupby([pd.Grouper(freq='Q'),"maille_nom"]).sum().sort_values(['date','maille_nom']).reset_index()
```

```python
total_by_month_by_department = df.groupby([pd.Grouper(freq='Q'),"maille_nom"]).sum().sort_values(['date','maille_nom']).reset_index()
```

Export des données sous format CSV pour pouvoir ensuite faire la jointure entre les deux tables 

```python
total_by_month_by_department.to_csv('Desktop/Ironhack/Projet1/cascovid.csv')
```
